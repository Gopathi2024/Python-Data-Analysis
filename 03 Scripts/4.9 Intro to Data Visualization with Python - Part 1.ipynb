# Part 1: Data Preparation and Combination

## Step 1: Import Libraries and Data
In this step, we will import the necessary libraries and the customer data set.


```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import os

# Path to project folder
project_path = r'C:\Users\sudee\OneDrive\Documents\Python Scripts\Instacart Basket Analysis'

# File paths
customer_data_path = os.path.join(project_path, '02 Data', 'Original Data', 'customers.csv')

# Importing customer data
df_customers = pd.read_csv(customer_data_path)

# Display the first few rows of the dataframe
df_customers.head()

```

## Step 2: Wrangle the Data
We will now clean the customer data to ensure it follows consistent logic.


```python
# Display column names and data types
df_customers.info()

# Rename columns for consistency
df_customers.rename(columns={'First Name': 'first_name', 'Surnam': 'surname', 'Age': 'age', 'Income': 'income'}, inplace=True)

# Drop columns that don't add value to the analysis
df_customers.drop(columns=['first_name'], inplace=True)

# Display the updated dataframe
df_customers.head()
```

## Step 3: Data Quality and Consistency Checks
We will perform data quality and consistency checks on the customer data.


```python
# Check for missing values
df_customers.isnull().sum()

# Fill or drop missing values if any
df_customers.dropna(inplace=True)

# Check for duplicates
df_customers.duplicated().sum()

# Drop duplicates if any
df_customers.drop_duplicates(inplace=True)

# Check for mixed-type data
df_customers.info()
```

## Step 4: Combine Customer Data with Instacart Data
We will now combine the customer data with the prepared Instacart data.


```python
# Path to the combined orders and products data
combined_data_path = os.path.join(project_path, '02 Data', 'Prepared Data', 'final_aggregated_orders_products.pkl')

# Load the combined orders and products data
df_combined = pd.read_pickle(combined_data_path)

# Check the structure of the combined dataframe
df_combined.info()

# Combine the dataframes on the 'user_id' column
df_combined = df_combined.merge(df_customers, on='user_id', how='left')

# Display the combined dataframe
df_combined.head()

```

## Step 5: Export the Combined DataFrame
We will export the combined dataframe for further analysis in Part 2.


```python
# Export the combined dataframe
combined_data_export_path = os.path.join(project_path, '02 Data', 'Prepared Data', 'combined_with_customers.pkl')
df_combined.to_pickle(combined_data_export_path)

# Confirm export
print(f"Combined DataFrame exported to {combined_data_export_path}")
```
